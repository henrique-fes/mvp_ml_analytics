## **1.Objetivo**:
Desenvolver um modelo de machine learning (classificação) que possa prever o diagnóstico de Doença de Alzheimer em pacientes, utilizando as informações levantadas por médicos durante as anamneses*,exames clínicos e históricos médicos anteriores. O modelo deve ser capaz de classificar os pacientes em duas categorias: diagnóstico positivo (1) ou diagnóstico negativo (0) para Alzheimer, com base em dados demográficos, fatores de estilo de vida, histórico médico, medições clínicas, e avaliações cognitivas e funcionais.

O diagnóstico de Alzheimer é um processo complexo e, muitas vezes, desafiador, devido à natureza progressiva e variável da doença. Além disso, os sintomas iniciais podem ser sutis e semelhantes a outras condições.

### **Vantagens de Usar Machine Learning no Diagnóstico de Alzheimer:**


1. Diagnóstico mais rápido e preciso, reduzindo o tempo e
o esforço dos médicos na análise de grandes volumes de dados.
2. Identificação de padrões subjacentes que podem não ser imediatamente evidentes, permitindo diagnóstico precoce.
3. Apoio à decisão clínica, auxiliando os médicos a tomar decisões mais informadas sobre o tratamento e acompanhamento do paciente.
4. Eficiência no uso de recursos em hospitais e clínicas, permitindo que os médicos se concentrem em casos mais complexos enquanto o modelo automatiza o diagnóstico para casos mais claros.


### **Resultados Esperados do MVP:**
- Um modelo funcional que classifique pacientes com base nos dados disponíveis, fornecendo previsões rápidas sobre o diagnóstico de Alzheimer.
- Validação de que as variáveis selecionadas podem ser usadas efetivamente para prever o diagnóstico da doença.
- Demonstração de valor para os profissionais de saúde, com potencial para integração em sistemas existentes.

**Informações do Paciente**  
**ID do Paciente**: Um identificador único atribuído a cada paciente (4751 a 6900).

**Age**: A idade dos pacientes varia de 60 a 90 anos.  
**Gender**: Gênero dos pacientes, onde 0 representa Masculino e 1 representa Feminino.  
**Ethnicity**: A etnia dos pacientes, codificada da seguinte forma:  
0: Caucasiano  
1: Afro-americano  
2: Asiático  
3: Outro  
**EducationLevel**: O nível de escolaridade dos pacientes, codificado da seguinte forma:  
0: Nenhuma  
1: Ensino Médio  
2: Bacharelado  
3: Pós-graduação  

**Fatores de Estilo de Vida**  
**BMI**: Índice de Massa Corporal dos pacientes, variando de 15 a 40.  
**Smoking**: Status de tabagismo, onde 0 indica Não e 1 indica Sim.  
**AlcoholConsumption**: Consumo semanal de álcool em unidades, variando de 0 a 20.  
**PhysicalActivity**: Atividade física semanal em horas, variando de 0 a 10.  
**DietQuality**: Pontuação de qualidade da dieta, variando de 0 a 10.  
**SleepQuality**: Pontuação de qualidade do sono, variando de 4 a 10.  

**Histórico Médico**  
**FamilyHistoryAlzheimers**: Histórico familiar de Doença de Alzheimer, onde 0 indica Não e 1 indica Sim.  
**CardiovascularDisease**: Presença de doença cardiovascular, onde 0 indica Não e 1 indica Sim.  
**Diabetes**: Presença de diabetes, onde 0 indica Não e 1 indica Sim.  
**Depression**: Presença de depressão, onde 0 indica Não e 1 indica Sim.  
**HeadInjury**: Histórico de lesão na cabeça, onde 0 indica Não e 1 indica Sim.  
**Hypertension**: Presença de hipertensão, onde 0 indica Não e 1 indica Sim.  

**Medições Clínicas**  
**SystolicBP**: Pressão arterial sistólica, variando de 90 a 180 mmHg.  
**DiastolicBP**: Pressão arterial diastólica, variando de 60 a 120 mmHg.  
**CholesterolTotal**: Níveis de colesterol total, variando de 150 a 300 mg/dL.  
**CholesterolLDL**: Níveis de colesterol lipoproteína de baixa densidade (LDL), variando de 50 a 200 mg/dL.  
**CholesterolHDL**: Níveis de colesterol lipoproteína de alta densidade (HDL), variando de 20 a 100 mg/dL.  
**CholesterolTriglycerides**: Níveis de triglicerídeos, variando de 50 a 400 mg/dL.  

**Avaliações Cognitivas e Funcionais**  
**MMSE**: Pontuação no Mini Exame do Estado Mental, variando de 0 a 30. Pontuações mais baixas indicam comprometimento cognitivo.  
**FunctionalAssessment**: Pontuação na avaliação funcional, variando de 0 a 10. Pontuações mais baixas indicam maior comprometimento.  
**MemoryComplaints**: Presença de queixas de memória, onde 0 indica Não e 1 indica Sim.  
**BehavioralProblems**: Presença de problemas comportamentais, onde 0 indica Não e 1 indica Sim.  
**ADL**: Pontuação das atividades de vida diária, variando de 0 a 10. Pontuações mais baixas indicam maior comprometimento.  

**Sintomas**  
**Confusion**: Presença de confusão, onde 0 indica Não e 1 indica Sim.  
**Disorientation**: Presença de desorientação, onde 0 indica Não e 1 indica Sim.  
**PersonalityChanges**: Presença de mudanças de personalidade, onde 0 indica Não e 1 indica Sim.  
**DifficultyCompletingTasks**: Presença de dificuldade em completar tarefas, onde 0 indica Não e 1 indica Sim.  
**Forgetfulness**: Presença de esquecimento, onde 0 indica Não e 1 indica Sim.  

**Informações de Diagnóstico**  
**Diagnosis**: Status do diagnóstico de Doença de Alzheimer, onde 0 indica Não e 1 indica Sim.  

**Informações Confidenciais**  
**DoctorInCharge**: Esta coluna contém informações confidenciais sobre o médico responsável, com "XXXConfid" como valor para todos os pacientes.

# **3.Modelagem e Treinamento**

Começaremos essa parte do trabalho importanto bibliotecas importantes para a modelagem, treinamento e classificação dos modelos pelos resultados.

```python
# Configuração para não exibir os warnings
import warnings
warnings.filterwarnings("ignore")

# Imports necessários
from sklearn.model_selection import KFold # para preparar os folds da validação cruzada
from sklearn.model_selection import cross_val_score # para executar a validação cruzada
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix# para a exibição das métricas do modelo
from sklearn.neighbors import KNeighborsClassifier # algoritmo KNN
from sklearn.tree import DecisionTreeClassifier # algoritmo Árvore de Classificação
from sklearn.linear_model import LogisticRegression # algoritmo Regressão Logística
from sklearn.naive_bayes import GaussianNB # algoritmo Naive Bayes
from sklearn.svm import SVC # algoritmo SVM
from sklearn.ensemble import RandomForestClassifier # algoritmo Random Forest
from sklearn.preprocessing import StandardScaler # para normalização dos dados
from sklearn.pipeline import Pipeline # para criar um pipeline de transformações e modelos
```

```python
# Modelagem

# Definindo uma seed global para esta célula de código
np.random.seed(7)

# Criando os folds para a validação cruzada
num_particoes = 10 # número de folds da validação cruzada
kfold = KFold(n_splits=num_particoes, shuffle=True, random_state=7) # faz o particionamento em 10 folds

# Listas para armazenar os modelos, os resultados e os nomes dos modelos
models = []
results = []
names = []

# Preparando os modelos e adicionando-os em uma lista
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('SVM', SVC()))
models.append(('NB', GaussianNB()))
models.append(('LR', LogisticRegression()))
models.append(('RFC', RandomForestClassifier()))


# Avaliando um modelo por vez
for name, model in models:
  cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='accuracy')
  results.append(cv_results)
  names.append(name)
  msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std()) # média e desvio padrão dos 10 resultados da validação cruzada
  print(msg)

# Boxplot de comparação dos modelos
fig = plt.figure()
fig.suptitle('Comparação da Acurácia dos Modelos')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

Visto que o RFC apresentou a maior acurácia, com um valor de 0.933523, indicando que ele é o modelo mais eficaz na previsão correta das classes em relação aos demais. Além disso, seu desvio padrão de 0.019906 é o menor entre todos os modelos avaliados, o que demonstra sua maior consistência e robustez, uma vez que o modelo é menos sensível a variações nos dados e oferece previsões mais estáveis.
Escolhemos o modelo e vamos avaliar como o modelo se comporta com os dados de teste e trazer mais métricas para corroborar a escolha. Em seguida tentaremos otimizar o modelo.

# **4. Avaliação de Resultados**


Avaliando o modelo nos dados de teste

```python
# Criando um modelo com todo o conjunto de treino
rfc = RandomForestClassifier(random_state= 7)
rfc.fit(X_train, y_train)

# Fazendo as predições com o conjunto de teste
predictions = rfc.predict(X_test)

# Estimando a acurácia no conjunto de teste
print(accuracy_score(y_test, predictions))
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

### 4.1 Otimização do Modelo

Para isso utilizaremos o RandomizedSearchCV.
O RandomizedSearchCV é uma técnica de otimização de hiperparâmetros usada para melhorar o desempenho de modelos de classificação (e regressão) em scikit-learn. Sua principal função é buscar de forma aleatória por uma combinação de hiperparâmetros que maximize o desempenho do modelo, sem testar todas as combinações possíveis, como no GridSearchCV. Em vez disso, ele amostra aleatoriamente um número definido de combinações de hiperparâmetros a partir de distribuições pré-definidas, permitindo uma busca mais eficiente, especialmente quando o espaço de hiperparâmetros é grande

```python
from sklearn.model_selection import RandomizedSearchCV
# Definir os parâmetros a serem testados
param_dist = {
    'n_estimators':[100, 200, 500],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5 ],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Realizar a busca aleatória
random_search = RandomizedSearchCV(estimator=rfc, param_distributions=param_dist, n_iter=100, cv=5, verbose=2, n_jobs=-1, random_state=7)

# Ajustar o modelo aos dados
random_search.fit(X_train, y_train)

# Melhor modelo encontrado
print("Melhores parâmetros:", random_search.best_params_)
```

```python
best_model = random_search.best_estimator_
predictions_bm = best_model.predict(X_test)
print(accuracy_score(y_test, predictions_bm))
print(classification_report(y_test,predictions_bm))
print(confusion_matrix(y_test,predictions_bm))
```
